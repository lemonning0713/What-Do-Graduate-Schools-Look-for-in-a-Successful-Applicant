```python
# Density plot for 2 groups
x0 = data[data["OfferOfAdmissionExtended"] == "YES"]["GPA"]
x1 = data[data["OfferOfAdmissionExtended"] == "NO"]["GPA"]

trace1 = go.Histogram(
    x=x0,
    opacity=0.75,
    name = 'Offer Of Admission Extended',
    text = "mean GPA: " + str(round(x0.mean(),2))
)
trace2 = go.Histogram(
    x=x1,
    opacity=0.75,
    name = "No",
    text = "mean GPA: " + str(round(x1.mean(),2))
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='overlay',
    title='GPA VS. Acceptance',
    xaxis=dict(
        range=[2.3, 4.4]
    ),
)
fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='overlaid histogram')
```
